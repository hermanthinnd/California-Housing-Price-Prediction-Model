```python
housing_na = housing.dropna(subset=['total_bedrooms'])
```

```python
housing_na.shape
```

```python
housing_na.info()
```

```python
dummies = pd.get_dummies(housing_na.ocean_proximity, dtype=int)
dummies
```

```python

housing_na_dummies = pd.concat([housing_na, dummies], axis='columns')
housing_na_dummies.head()       
```

```python
housing_clean = housing_na_dummies.drop(['ocean_proximity', 'ISLAND'], axis='columns')
housing_clean
```

```python
X=housing_clean.drop('median_house_value', axis='columns')
X.head()
y=housing_clean['median_house_value']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1984) 
```

```python
from sklearn.linear_model import LinearRegression
OLS = LinearRegression()
OLS.fit(X_train, y_train)
```

```python
print('Intercept:', OLS.intercept_)
print('Coefficients:', OLS.coef_)
print('R-Squared:', OLS.score(X_test, y_test))
```

```python
#Predicting with OLS
y_pred = OLS.predict(X_test)
performance = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
performance['Error'] = performance['Actual'] - performance['Predicted']
performance.head()
```

```python
#preparing data for visualization
performance.reset_index(drop=True, inplace=True)  
performance.reset_index(inplace=True)
performance.head()
```

```python
#plotting the residuals
plt.figure(figsize=(10,6))  
plt.bar('index','Error', data=performance[:50],  color='blue', width=0.3)
plt.xlabel('Observations')
plt.ylabel('Residuals')
plt.title('Residual Plot')
plt.show()
```

```python
import statsmodels.api as sm
X_train_sm = sm.add_constant(X_train)
X_train_sm.head()
OLS_sm = sm.OLS(y_train, X_train_sm).fit()
OLS_sm.summary()
```

```python
plt.scatter(performance['Actual'], performance['Predicted'], alpha=0.5)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values')
plt.show()
```
